```python
import pandas as pd

data2 = {
    "Duration": [60, 450, 30],
    "Date": ["2020/12/01", "2020/12/08", "2020/12/09"],
    "Pulse": [110, 104, 109]
}

df = pd.DataFrame(data2)

print("Before:")
print(df)

for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120

print("\nAfter:")
print(df)
```
